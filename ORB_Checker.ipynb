```python
import os
import eikon as ek
import numpy as np
import pandas as pd
from matplotlib import pyplot
import time
import warnings
warnings.filterwarnings("ignore")


import datetime

#############################
symbolss= ['AVEU.NS','BRGR.NS','NEST.NS','PIDI.NS','HDFL.NS','PROC.NS','ASPN.NS','HLL.NS','UNSP.NS','TITN.NS','BION.NS','BRIT.NS','DABU.NS','PAGE.NS','SBIL.NS',
           'ICIR.NS','SHCM.NS','HVEL.NS','ICIL.NS','UBBW.NS','DIVI.NS','BOSH.NS','HDFA.NS','COLG.NS','REDY.NS','MRCO.NS','THEE.NS','SIEM.NS','CIPL.NS','SUN.NS',
           'MRTI.NS','KTKM.NS','GOCP.NS','DLF.NS','JSTL.NS','AXBK.NS','BJFN.NS','MOSS.NS','ADAI.NS','RELI.NS','BJFS.NS','ICBK.NS','GRAS.NS','BPCL.NS','TCS.NS',
           'EICH.NS','HDBK.NS','CCRI.NS','CADI.NS','ITC.NS','ULTC.NS','ABUJ.NS','UPLL.NS','APSE.NS','INFY.NS','ACC.NS','PLNG.NS','ARBN.NS','BAJA.NS','LART.NS',
           'ORCL.NS','HDFC.NS','BANH.NS','HROM.NS','HCLT.NS','BHRI.NS','WIPR.NS','BOB.NS','TEML.NS','ZEE.NS','HZNC.NS','MAHM.NS','GENA.NS','BJAT.NS','PGRD.NS',
           'NHPC.NS','NTPC.NS','ASOK.NS','SBI.NS','GAIL.NS','INBK.NS','HALC.NS','NMDC.NS','LTFH.NS','SRTR.NS','VDAN.NS','COAL.NS','HPCL.NS','IOC.NS','TISC.NS',
           'ONGC.NS','PWFC.NS','INBF.NS','BRTI.NS','INGL.NS','LUPN.NS','TAMO.NS','TAMdv.NS','PNBK.NS','VODA.NS','PIRA.NS']
ek.set_app_key('APP_KEY')

```

```python

while(True):
    stt_date = (datetime.datetime.now().strftime("%Y-%m-%d 03:45:00"))
    end_date = ((datetime.datetime.now() - datetime.timedelta(minutes=330)).strftime("%Y-%m-%d %H:%M:%S"))
    OR_Buys = []
    OR_Sells = []


    vols = ek.get_timeseries(rics=symbolss,fields=['VOLUME'],start_date=stt_date,end_date=end_date,interval='hour',corax='adjusted')
    factor = 3600/((datetime.datetime.now() - datetime.timedelta(minutes=330)) - vols.index.tolist()[-1]).seconds
    vols[-1:] = vols[-1:]*factor

    vols = vols - vols.shift(1)
    vols[vols<0] = 0

    time.sleep(2)
    closes = ek.get_timeseries(rics=symbolss,fields=['CLOSE'],start_date=stt_date,end_date=end_date,interval='hour',corax='adjusted')
    closes = ((closes - closes.shift(1))/closes.shift(1))
    syms = (vols*closes).transpose().sum(axis=1).sort_values(ascending=False)

#     syms = ((vols*closes)[-1:].transpose().iloc[:,0] - (vols*closes)[-2:-1].transpose().iloc[:,0]).sort_values(ascending=False)

    syms = syms[:5].index.tolist() + syms[-5:].index.tolist()
    # time.sleep(2)

    for sym in syms:

        lst = ek.get_timeseries(rics=[sym],fields=['HIGH','LOW'],start_date=stt_date,end_date=end_date,interval='minute',corax='adjusted')

    #     lst['HIGH'] = lst['HIGH'].resample('5Min').ohlc()['high']
    #     lst['LOW'] = lst['LOW'].resample('5Min').ohlc()['low']

        OR_max = (lst['HIGH'][:60].max())
        OR_min = (lst['LOW'][:60].min())
        Now_max = (lst['LOW'][:-30].max())
        Now_min = (lst['HIGH'][:-30].min())

        if(Now_max > OR_max):
            OR_Buys.append(sym)

        elif(Now_min < OR_min):
            OR_Sells.append(sym)

        time.sleep(3)

    print("Buys:")
    print(OR_Buys)

    print("Sells:")
    print(OR_Sells)
    print("===============================================================")
    print(" ")
    os.system("say 'Published'")

    time.sleep(300)
```
